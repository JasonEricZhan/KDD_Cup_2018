```python
#from https://github.com/mapado/haversine/blob/master/haversine/__init__.py
from math import radians, cos, sin, asin, sqrt

AVG_EARTH_RADIUS = 6371  # in km
MILES_PER_KILOMETER = 0.621371
from math import radians, cos, sin, asin, sqrt

AVG_EARTH_RADIUS = 6371  # in km
MILES_PER_KILOMETER = 0.621371

def haversine(point1, point2, miles=False):
    """ Calculate the great-circle distance between two points on the Earth surface.
    :input: two 2-tuples, containing the latitude and longitude of each point
    in decimal degrees.
    Example: haversine((45.7597, 4.8422), (48.8567, 2.3508))
    :output: Returns the distance between the two points.
    The default unit is kilometers. Miles can be returned
    if the ``miles`` parameter is set to True.
    """
    # unpack latitude/longitude
    lat1, lng1 = point1
    lat2, lng2 = point2

    # convert all latitudes/longitudes from decimal degrees to radians
    lat1, lng1, lat2, lng2 = map(radians, (lat1, lng1, lat2, lng2))

    # calculate haversine
    lat = lat2 - lat1
    lng = lng2 - lng1
    d = sin(lat * 0.5) ** 2 + cos(lat1) * cos(lat2) * sin(lng * 0.5) ** 2
    h = 2 * AVG_EARTH_RADIUS * asin(sqrt(d))
    if miles:
        return h * MILES_PER_KILOMETER # in miles
    else:
        return h  # in kilometers

    

```

### Mapping from AQI to grid station, by navie intuition:  
All distance mutiply -1 to get the rank, and muptily to the weather matrix in grid, to be faster, I choose first 150 closest grid point

```python
dist_AQItoGrid_Rank=OrderedDict()
for keys in dist_AQItoGrid:
    dist_AQItoGrid_Rank[keys]=-(dist_AQItoGrid[keys])
```

```python
import time
start=time.time()
AQItoWeather_matrix=OrderedDict()
for domain_key in dist_AQItoGrid_Rank:
    rank_vector=dist_AQItoGrid_Rank[domain_key]
    dist_vector=dist_AQItoGrid[domain_key]
    ith=0
    target_index_set=dist_vector.argsort()[:150]
    for codmain_key in weather_Grid:
        index=mapping_toGrid[codmain_key]
        if index not in target_index_set:
            continue
        weight=rank_vector[index]
        if ith is 0:
            container=weather_Grid[codmain_key]*weight
        else:
            weather_matrix=weather_Grid[codmain_key]*weight
            container=np.vstack((weather_matrix,container))
        ith+=1
    AQItoWeather_matrix[domain_key]=container
    del container
end=time.time()
print("time: "+str(end-start))
```
