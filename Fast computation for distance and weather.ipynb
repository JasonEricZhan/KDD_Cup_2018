```python
#load position
beijingAqCsv_sta_position = pd.read_csv('Beijing_AirQuality_Stations.csv')
#preparing
air_station_beij=beji_aqi_summary.groupby("stationId", sort=False) 
```

```python
location_AQI_beijin={}
for i in range(0,len(beijingAqCsv_sta_position)):
    temp=beijingAqCsv_sta_position.iloc[i,:]
    if temp['Pollutant Species'] in list(air_station_beij.groups.keys()):
       location_AQI_beijin[temp['Pollutant Species']]=temp.values[1:].astype(float).tolist()[::-1]
```

```python
#from https://github.com/mapado/haversine/blob/master/haversine/__init__.py
from math import radians, cos, sin, asin, sqrt

AVG_EARTH_RADIUS = 6371  # in km
MILES_PER_KILOMETER = 0.621371
def haversine(point1, point2, miles=False):
    """ Calculate the great-circle distance bewteen two points on the Earth surface.

    :input: two 2-tuples, containing the latitude and longitude of each point
    in decimal degrees.

    Example: haversine((45.7597, 4.8422), (48.8567, 2.3508))

    :output: Returns the distance bewteen the two points.
    The default unit is kilometers. Miles can be returned
    if the ``miles`` parameter is set to True.

    """
    # unpack latitude/longitude
    lat1, lng1 = point1
    lat2, lng2 = point2

    # convert all latitudes/longitudes from decimal degrees to radians
    lat1, lng1, lat2, lng2 = map(radians, (lat1, lng1, lat2, lng2))

    # calculate haversine
    lat = lat2 - lat1
    lng = lng2 - lng1
    d = sin(lat * 0.5) ** 2 + cos(lat1) * cos(lat2) * sin(lng * 0.5) ** 2
    h = 2 * AVG_EARTH_RADIUS * asin(sqrt(d))
    if miles:
        return h * 0.621371  # in miles
    else:
        return h  # in kilometers
    

```

```python
#35X651
dist_AQItoGrid={}
for key_1 in location_AQI_beijin:
    distance_vector=np.array([])
    point1=location_AQI_beijin[key_1]
    for key_2 in location_grid_beijin:
        point2=location_grid_beijin[key_2]
        dist=haversine(point1,point2, miles=False)
        distance_vector=np.append(distance_vector,dist)
    dist_AQItoGrid[key_1]=distance_vector
```
